```python
# Sajjad Shamsnia
# Problem 2
data = []
f = open('ex1_data.txt',mode = 'rt') 

for line in f:
    line = line.replace('\n','')
    data.append(line)

data_mod = []
for member in data:
    if member == '':
        data_mod.append(True)
    else:
        data_mod.append(int(member))
```

```python
# Part A
data_ave = []
ave_list = []
for i in data_mod:
    if type(i) == int:
        data_ave.append(i)
    elif i == True:
        ave = sum(data_ave)/len(data_ave)
        ave = round(ave,2)
        ave_list.append(ave)
        data_ave = []

print('average value list = ', ave_list)
print('maximum value in averages = ', max(ave_list))
```

```python
# Part B
three_highest = []
three_highest.append(max(data_ave))
data_ave.remove(max(data_ave))
three_highest.append(max(data_ave))
data_ave.remove(max(data_ave))
three_highest.append(max(data_ave))
data_ave.remove(max(data_ave))
print('The average of the three highest values = ', sum(three_highest)/len(three_highest))
```
